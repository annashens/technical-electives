```python
import pandas as pd
```

```python
courses_df =pd.read_csv('courses.csv', index_col = 'course_code')
courses_df['liked'] = courses_df['liked'].str.slice(0, -1)
courses_df['easy'] = courses_df['easy'].str.slice(0, -1).astype(int)
courses_df['useful'] = courses_df['useful'].str.slice(0, -1)
courses_df.sort_values(by='easy', ascending=False, inplace=True)
courses_df
```

```python
from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.charts import Liquid
from pyecharts.globals import SymbolType
from pyecharts.charts import Bar, Line
bar = (
 Bar(init_opts=opts.InitOpts(height="400px", width="3000px"))
 .add_xaxis(courses_df.index.tolist())
 .add_yaxis('easy', courses_df['easy'].to_list(),
            bar_width= '10%')
 .add_yaxis('liked', courses_df['liked'].to_list(),
            bar_width= '10%'
            )
 .add_yaxis('useful', courses_df['useful'].to_list(),
            category_gap='150%',
            bar_width= '10%')
 .extend_axis(
        yaxis=opts.AxisOpts(
            name="Percentage",
            type_="value",
            min_=0,
            max_=100,
            position="right",
            axisline_opts=opts.AxisLineOpts(
                linestyle_opts=opts.LineStyleOpts(width=2)
            )
        )
 )
 .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False),
        # markpoint_opts=opts.MarkPointOpts(
        # data=[opts.MarkPointItem(type_="max",value=20, name="Highest"),
        #       opts.MarkPointItem(type_="min", name="Lowest")]
		# 													),
    )
 .set_global_opts(
     xaxis_opts=opts.AxisOpts(name="Category", axislabel_opts=opts.LabelOpts(interval=0, rotate=45)),
    #  title_opts=opts.TitleOpts(title="Technical Electives")
 )
)


bar.load_javascript()
bar.render_notebook()
bar.render("bar_chart.html")
```
